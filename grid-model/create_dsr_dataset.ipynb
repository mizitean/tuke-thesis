```python
import pandas as pd
from pathlib import Path

# Path to the folder containing PEMMDB xlsx files
PEMMDB_DIR = Path("../data/tyndp2024/PEMMDB2/2030")
DSR_SHEET = "DSR"
ROWS_TO_READ = 9
SKIP_HEADER_ROWS = 6   # skip first 6 rows (non-data header rows)
SKIP_TAIL_ROWS = 2     # skip last 2 rows

all_dfs = []

for xlsx_file in sorted(PEMMDB_DIR.glob("*.xlsx")):
    # Extract country/node code from filename, e.g. PEMMDB_SK00_NationalTrends_2030.xlsx -> SK00
    node = xlsx_file.stem.split("_")[1]
    
    try:
        df = pd.read_excel(xlsx_file, sheet_name=DSR_SHEET, nrows=ROWS_TO_READ, header=SKIP_HEADER_ROWS)
        # Drop last 2 rows
        df = df.iloc[:-SKIP_TAIL_ROWS]
        df.insert(0, "node", node)          # add source file identifier
        df.insert(1, "source_file", xlsx_file.name)
        all_dfs.append(df)
        print(f"✓ {xlsx_file.name}")
    except Exception as e:
        print(f"✗ {xlsx_file.name}: {e}")

# Combine all results into a single DataFrame
dsr_df = pd.concat(all_dfs, ignore_index=True) if all_dfs else pd.DataFrame()
print(f"\nTotal rows collected: {len(dsr_df)}")
```

```python
# Transpose DSR data: Price Bands become rows, attributes become columns
#
# In dsr_df each row is one attribute (Capacity, Units, Hours, Price, ...),
# the 4th column (index 3 after node/source_file) holds the attribute name,
# and the remaining columns hold values per Price Band.
#
# We want one row per (node, price_band) with columns: Capacity, Units, Hours, Price, ...

ATTRIBUTE_COL = 3   # positional index of the column holding the attribute name (after node, source_file, unnamed)

all_transposed = []

for node, group in dsr_df.groupby("node"):
    group = group.reset_index(drop=True)

    # The attribute name column is the 4th column (index 3)
    attr_col = group.columns[ATTRIBUTE_COL]

    # Row 0 of this group contains Price Band names
    price_bands_raw = group.iloc[0, ATTRIBUTE_COL + 1:].tolist()

    # Deduplicate price band names (e.g. "Price Band 14" appears 3x for different climate years)
    seen = {}
    unique_bands = []
    for name in price_bands_raw:
        if pd.isna(name):
            unique_bands.append(name)
            continue
        if name in seen:
            seen[name] += 1
            unique_bands.append(f"{name}_{seen[name]}")
        else:
            seen[name] = 1
            unique_bands.append(name)

    # Remaining rows are the actual attribute rows (Capacity, Units, Hours, Price, ...)
    attr_df = group.iloc[1:].copy()
    attr_df = attr_df.set_index(attr_col)
    attr_df.index = attr_df.index.astype(str)
    attr_df.index.name = None
    attr_df = attr_df.iloc[:, ATTRIBUTE_COL:]   # drop node, source_file, unnamed col
    attr_df.columns = range(len(attr_df.columns))
    attr_df.columns = unique_bands

    # Drop NaN separator columns
    attr_df = attr_df[[c for c in attr_df.columns if not (isinstance(c, float) and pd.isna(c))]]

    # Transpose: price bands → rows, attributes → columns
    transposed = attr_df.T.copy()
    transposed.index.name = "price_band"
    transposed = transposed.reset_index()

    # Ensure all column names are unique strings
    cols = list(transposed.columns)
    seen_cols = {}
    unique_cols = []
    for c in cols:
        c_str = str(c)
        if c_str in seen_cols:
            seen_cols[c_str] += 1
            unique_cols.append(f"{c_str}_{seen_cols[c_str]}")
        else:
            seen_cols[c_str] = 1
            unique_cols.append(c_str)
    transposed.columns = unique_cols

    # Add node identifier
    transposed.insert(0, "node", node)

    all_transposed.append(transposed)

# Combine all nodes — use outer join so nodes with different numbers of attributes still merge
dsr_transposed = pd.concat(all_transposed, ignore_index=True, join="outer")
print(f"Total rows: {len(dsr_transposed)}")
print(dsr_transposed.shape)
```

```python
dsr_transposed["Capacity"] = pd.to_numeric(dsr_transposed["Capacity"], errors="coerce")
dsr_transposed = dsr_transposed.dropna(subset=["Capacity"])

# Drop NaN padding columns produced by pd.concat outer join
nan_cols = ["nan", "nan_2", "nan_3", "nan_4", "nan_5", "nan_6"]
dsr_transposed = dsr_transposed.drop(columns=[c for c in nan_cols if c in dsr_transposed.columns])

print(f"Rows after dropping NaN Capacity: {len(dsr_transposed)}")
print(f"Columns: {list(dsr_transposed.columns)}")
print(len(dsr_transposed))
display(dsr_transposed.head(100))
```

## Parse Timeseries
- p_max_pu

```python

import pandas as pd
from pathlib import Path

PEMMDB_DIR = Path("../data/tyndp2024/PEMMDB2/2030")
DSR_SHEET = "DSR"
OUTPUT_DIR = Path("../data/open-tyndp")
CLIMATE_YEAR_TS = 2009

# Build a year-2030 date range (8760 hours, hour 1 = 00:00)
timestamps = pd.date_range("2030-01-01 00:00", periods=8760, freq="h")

all_cols = {}

for xlsx_file in sorted(PEMMDB_DIR.glob("*.xlsx")):
    node = xlsx_file.stem.split("_")[1]

    try:
        raw = pd.read_excel(xlsx_file, sheet_name=DSR_SHEET, header=None)
    except Exception as e:
        print(f"✗ {xlsx_file.name}: {e}")
        continue

    # ── 1. Read static metadata (rows 7-13) ──────────────────────────────────
    band_names_raw = raw.iloc[7,  2:].tolist()
    capacities_raw = raw.iloc[8,  2:].tolist()
    cy_start_raw   = raw.iloc[12, 2:].tolist()
    cy_end_raw     = raw.iloc[13, 2:].tolist()

    # ── 2. Deduplicate band names globally (matching dsr_pemmdb.csv logic) ───
    seen = {}
    deduped_names = []
    for name in band_names_raw:
        if pd.isna(name):
            deduped_names.append(None)
            continue
        name_str = str(name).strip()
        if name_str in seen:
            seen[name_str] += 1
            deduped_names.append(f"{name_str}_{seen[name_str]}")
        else:
            seen[name_str] = 1
            deduped_names.append(name_str)

    # ── 3. Filter: capacity > 0 AND climate year start <= 2009 <= climate year end ──
    labeled = []
    for i, (band, cap, cy_s, cy_e) in enumerate(zip(deduped_names, capacities_raw, cy_start_raw, cy_end_raw)):
        if band is None:
            continue
        try:
            cap  = float(cap)
            cy_s = int(float(cy_s))
            cy_e = int(float(cy_e))
        except (ValueError, TypeError):
            continue
        if pd.isna(cap) or cap <= 0:
            continue
        if not (cy_s <= CLIMATE_YEAR_TS <= cy_e):
            continue
        labeled.append((i, band, cap))

    if not labeled:
        print(f"  {node}: no bands for CY{CLIMATE_YEAR_TS}, skipping")
        continue

    # ── 4. Extract timeseries rows (row 15 onwards = index 15, 8760 rows) ────
    ts_data = raw.iloc[15:15 + 8760].reset_index(drop=True)

    for raw_col_idx, band, capacity in labeled:
        col_key = f"{node}_{band}"
        ts_col = ts_data.iloc[:, raw_col_idx + 2]   # +2 for Date and Hour offset
        ts_values = pd.to_numeric(ts_col, errors="coerce").fillna(0.0)
        # p_max_pu = available_MW / capacity_MW  (clipped to [0, 1])
        p_max_pu = (ts_values / capacity).clip(0.0, 1.0)
        all_cols[col_key] = p_max_pu.values

    print(f"✓ {node}: {len(labeled)} bands")

# ── 5. Build final DataFrame ─────────────────────────────────────────────────
dsr_timeseries = pd.DataFrame(all_cols, index=timestamps)
dsr_timeseries.index.name = "timestamp"

print(f"\nTimeseries shape: {dsr_timeseries.shape}")
de_cols = [c for c in dsr_timeseries.columns if c.startswith("DE00_")]
print(f"DE00 bands ({len(de_cols)}): {[c.replace('DE00_', '') for c in de_cols]}")
dsr_timeseries.head()
```

```python

OUTPUT_DIR.mkdir(parents=True, exist_ok=True)
out_path = OUTPUT_DIR / "dsr_p_max_pu_timeseries.csv"
dsr_timeseries.to_csv(out_path)
print(f"Exported → {out_path}")
print(f"Shape: {dsr_timeseries.shape}")
```
